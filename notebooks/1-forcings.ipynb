```python
g= xr.open_zarr('../data/training_data.zarr/')
```

```python
import uwnet.thermo

def compute_advection_tedency_xarray(g, key):
    
    # handle grid information
    dz = (uwnet.thermo.get_dz(g.z)).values
    rho = (g.layer_mass / dz).values
    dz, rho = [np.reshape(x, (-1, 1,1)) for x in [dz, rho] ]
    dx = float(g.x[1] -g.x[0])    
    
    # handle 4D inputs
    dim_order = ['time', 'z', 'y', 'x']
    u,v,w,f = [g[key].transpose(*dim_order).values for key in ['U', 'V', 'W', key]]
    
    out = c_grid_advective_tendency(u, v, w, f, dx, dz, rho)
    
    out = xr.DataArray(out, dims=g.U.dims, coords=g.U.coords)
    out.attrs['units'] = g[key].attrs.get('units', '1') + '/s'
    return out

forcings = xr.Dataset(dict(
    FQT = tend(g.isel(time=slice(0,1)), 'qt'),
    FSL = tend(g.isel(time=slice(0,1)), 'sl')))
```
