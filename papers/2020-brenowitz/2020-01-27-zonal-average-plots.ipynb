```python
import xarray as xr
%matplotlib inline
import uwnet.thermo
```

```python
ds = xr.open_dataset("../../data/processed/training.mean.nc")
```

```python
ds['Q'] = uwnet.thermo.midtropospheric_moisture(ds.QV, ds.p)
ds['LTS'] = uwnet.thermo.lower_tropospheric_stability(ds.TABS, ds.p, ds.SST)
ds['lat'] = uwnet.thermo.ngaqua_y_to_lat(ds.y)
```

```python
df = ds[['SST', 'Q', 'LTS', 'lat']].to_dataframe().set_index('lat')
```

```python
df.plot(subplots=True, figsize=(10, 10), color='blue', grid=True)
```
